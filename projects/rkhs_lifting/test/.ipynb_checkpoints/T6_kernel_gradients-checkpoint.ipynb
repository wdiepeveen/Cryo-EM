```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from projects.rkhs_lifting.src.integrators.base.sd60 import SD60
from projects.rkhs_lifting.src.integrators.multi.local_regular import Local_Regular

# TODO get some random rotations
base_integrator = SD60()
# TODO test local_regular grid integrator
multi_integrator = Local_Regular(quaternions=base_integrator.quaternions,l=3,sep_dist=np.pi/11)

# print(multi_integrator.angles.shape)
quat_ind = 10
orig_quat = base_integrator.quaternions[quat_ind,:]
quats = multi_integrator.quaternions[:,quat_ind,:]  # Get quaternions for starting rotation x

print(orig_quat.shape)
print(quats)
```

```python
from projects.rkhs_lifting.src.kernels.rescaled_cosine import Rescaled_Cosine_Kernel

kernel = Rescaled_Cosine_Kernel(quaternions=quats)

gradients_at_orig_quat = kernel.gradient(free_quaternion=orig_quat[None,None,:],fixed_quaternion=quats[None,:,:])

print(gradients_at_orig_quat)
print(gradients_at_orig_quat.shape)
```

```python
print(base_integrator.quaternions[:,None,:].shape)
print(multi_integrator.quaternions.swapaxes(0,1).shape)

gradients_at_orig_quats = kernel.gradient(free_quaternion=base_integrator.quaternions[:,None,:],
                                          fixed_quaternion=multi_integrator.quaternions.swapaxes(0,1))

print(gradients_at_orig_quats.shape)

# We should get here a (n,N,4) array (l^3,60,4)
```

```python
# TODO plot original rots
orig_angles = base_integrator.angles

x = orig_angles[:, 0]
y = orig_angles[:, 1]
z = orig_angles[:, 2]
print("Number of angles = {}".format(len(x)))

fig = plt.figure()
ax = fig.gca(projection='3d')

img = ax.scatter(x, y, z)
ax.set_xlabel("$\phi$")
ax.set_ylabel("$\\theta$")
ax.set_zlabel("$\psi$")

# TODO plot local grid in separate plot
n = multi_integrator.n
N = multi_integrator.N
local_angles = multi_integrator.angles.reshape((n*N,3))

x2 = local_angles[:, 0]
y2 = local_angles[:, 1]
z2 = local_angles[:, 2]
print("Number of angles = {}".format(len(x2)))

fig = plt.figure()
ax = fig.gca(projection='3d')

img = ax.scatter(x2, y2, z2)
ax.set_xlabel("$\phi$")
ax.set_ylabel("$\\theta$")
ax.set_zlabel("$\psi$")
```
