```python
import numpy as np
import os
import logging

from experiments.testing_joint_refinement.E1_run import run_experiment
from experiments.testing_joint_refinement.E2_post_processing import post_processing

from tools.exp_tools import Exp

logger = logging.getLogger(__name__)
```

```python
# Experiment parameters
vol_smudge = 10
mr_repeat = 2 # number of times mesh refinement
snr = 1 / 16  
num_imgs = 2048
J0 = 15

max_iter = 10

# Directly start experiment
exp = Exp()

exp.begin(prefix="exp_joint_refinement" , postfix="SNR{}_N{}_J{}_r{}_s{}_i{}".format(int(1/snr), num_imgs, J0, mr_repeat, vol_smudge, max_iter))
exp.dbglevel(4)
```

```python
# Set data path
data_dir = "data"
# data_filename = "clean70SRibosome_vol_65p.mrc"
data_filename = "waving_spike_state22_91p.mrc"
data_path = os.path.join("..", "..",data_dir, data_filename)

# Set RELION results path
RELION_folder = "RELION_run_rotations"
```

```python
# Experiment

logger.info("Start Experiment")
logger.info("Running Experiment for SNR = {} and {} images".format(snr, num_imgs))

run_experiment(exp=exp,
               max_iter=max_iter,
               num_imgs=num_imgs,
               snr=snr,
               mr_repeat=mr_repeat,
               J0=J0,
               data_path=data_path,
               vol_smudge=vol_smudge,
               )
```

```python
logger.info("Start Postprocessing")
# Postprocessing
post_processing(exp=exp,
                mr_repeat=mr_repeat,
                hist_drange=20,
                hist_dvrange=400,
                hist_Jrange=20,
                hist_Jvrange=600,
                RELION_folder=RELION_folder,
                )
```
